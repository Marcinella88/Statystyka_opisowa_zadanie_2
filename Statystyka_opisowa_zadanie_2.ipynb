```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data_1 = np.random.normal(loc=3, scale=1, size=100)
data_2 = np.random.normal(loc=3, scale=1, size=10000)
```

**1. ŚREDNIA:**

```python
mean_data_1 = np.round(np.mean(data_1),4)
mean_data_2 = np.round(np.mean(data_2),4)

print(f"Średnia dla zbioru 'data_1' wynosi: {mean_data_1} .")
print(f"Średnia dla zbioru 'data_2' wynosi: {mean_data_2} .")
```

Średnia charakteryzuje się tym, że bieże pod uwagi również skrajne wartości które mogą znacznie zaburzyć jej wynik w przypadku nie wielkiej ilości danych. Średnią najlepiej używać w sytuacji kiedy w zbiorze danych brakuje wartości odstających.

**2. MEDIANA:**

```python
median_data_1 = np.round(np.median(data_1),4)
median_data_2 = np.round(np.median(data_2),4)

print(f"Mediana dla zbioru 'data_1' wynosi: {median_data_1} .")
print(f"Mediana dla zbioru 'data_2' wynosi: {median_data_2} .")
```

Mediana charakteryzuje się tym, że pomija skrajne wartości, które mogą zaburzyć wynik.

**3. MODA:**

3.1 Tu wprowadzam zmianę, aby wartości były zaokrąglana do najbliższej liczby całkowitej. W innym wypadku moda pokazuje wartości de facto unikalne to wynika z dużej dokładności generowania liczb.

```python
data_1_int = np.round(data_1).astype(int) # zaokrąglamy do najbliższej liczby całkowitej
data_2_int = np.round(data_2).astype(int) # zaokrąglamy do najbliższej liczby całkowitej

mode_data_1_int = stats.mode(data_1_int, keepdims=True)
mode_data_1_int_value = mode_data_1_int.mode[0]
mode_data_1_int_count = mode_data_1_int.count[0]

mode_data_2_int = stats.mode(data_2_int, keepdims=True)
mode_data_2_int_value = mode_data_2_int.mode[0]
mode_data_2_int_count = mode_data_2_int.count[0]

print(f"Moda dla zbioru 'data_1' wynosi: {mode_data_1_int_value} - wystąpiła: {mode_data_1_int_count} razy.")
print(f"Moda dla zbioru 'data_2' wynosi: {mode_data_2_int_value} - wystąpiła: {mode_data_2_int_count} razy.")
```

Moda podaje najczęściej powtarzającą się wartość oraz ilość - ile razy najpopularniejsza wartość się powtórzyła w zbiorze.

**4. KWARTYLE:**

```python
q0_d1 = np.quantile(data_1, 0.0).round(4)
q1_d1 = np.quantile(data_1, 0.25).round(4)
q2_d1 = np.quantile(data_1, 0.5).round(4)
q3_d1 = np.quantile(data_1, 0.75).round(4)
q4_d1 = np.quantile(data_1, 1.0).round(4)

q0_d2 = np.quantile(data_2, 0.0).round(4)
q1_d2 = np.quantile(data_2, 0.25).round(4)
q2_d2 = np.quantile(data_2, 0.5).round(4)
q3_d2 = np.quantile(data_2, 0.75).round(4)
q4_d2 = np.quantile(data_2, 1.0).round(4)

print(f"Rozkład kwartyli dla data_1: Q0 - {q0_d1}, Q1 - {q1_d1}, Q2 - {q2_d1}, Q3 - {q3_d1}, Q4 - {q4_d1}")
print(f"Rozkład kwartyli dla data_2: Q0 - {q0_d2}, Q1 - {q1_d2}, Q2 - {q2_d2}, Q3 - {q3_d2}, Q4 - {q4_d2}")
```

```python
plt.hist(data_1, bins=20)
plt.axvline(x = mean_data_1, color='red', label = 'Średnia')
plt.axvline(x = median_data_1, color='green', label = 'Mediana')
plt.axvline(x = mode_data_1_int_value, color='orange', label = 'Moda')
plt.axvline(x = q0_d1, color='black', label = 'Min')
plt.axvline(x = q1_d1, color='black', label = 'Q1')
plt.axvline(x = q3_d1, color='black', label = 'Q3')
plt.axvline(x = q4_d1, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Data 100")
plt.show()

plt.hist(data_2, bins=20)
plt.axvline(x = mean_data_2, color='red', label = 'Średnia')
plt.axvline(x = median_data_2, color='green', label = 'Mediana')
plt.axvline(x = mode_data_2_int_value, color='orange', label = 'Moda')
plt.axvline(x = q0_d2, color='black', label = 'Min')
plt.axvline(x = q1_d2, color='black', label = 'Q1')
plt.axvline(x = q3_d2, color='black', label = 'Q3')
plt.axvline(x = q4_d2, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Data 10000")
plt.show()
```

Kwartyle dzielą zbiór danych na cztery części gdzie Q0 to wartość minimalna, Q1 - kwartyl pierwszy to wartość graniczna gdzie w zbiorze jest 25% wartości mniejszych i 75% większych. Q2 to mediana - dzieli zbiór na dwie równe części pod względem ilości obserwacji. Q3 - kwartyl pierwszy to wartość graniczna gdzie w zbiorze jest 75% wartości mniejszych i 25% większych. Q4 - to wartość maksymalna.

**5. ZAKRES:**

```python
range_data_1 = max(data_1) - min(data_1)
range_data_2 = max(data_2) - min(data_2)

print(f"Zakres danych zbioru data_1 wynosi: {range_data_1}")
print(f"Zakres danych zbioru data_2 wynosi: {range_data_2}")
```

Pierwsza metryka związana z rozproszeniem, czyli zmiennością rozkładu. Zakres to rozpiętość danych od wartości najniższej do najwyższej. Duży przedział oznacz dużą zmienność, a mały przedział oznacza małą zmienność.

**6. ROZSTĘP MIĘDZYKWARTYLOWY = IQR:**

```python
IQR_d1 = q3_d1 - q1_d1
IQR_d2 = q3_d2 - q1_d2

print(f"Rozstęp międzykwartylowy dla data_1 wynosi: {IQR_d1}")
print(f"Rozstęp międzykwartylowy dla data_2 wynosi: {IQR_d2}")
```

IQR zawiera pierwszy i trzeci kwartyl, jest to połowa rozkładu zmiennej. Na jej bazie jest tworzony wykres pudełkowy - BoxPlot.

**7. WARIANCJA:**

```python
variance_data_1 = np.var(data_1, ddof=1)
variance_data_2 = np.var(data_2, ddof=1)

print(f"Wariancja dla data_1 wynosi: {variance_data_1}")
print(f"Wariancja dla data_2 wynosi: {variance_data_2}")
```

Wariancja to miara rozpiętości między liczbami w zbiorze danych. Wariancja mierzy jak bardzo każda liczba w zbiorze jest oddalona od średniej.

**8. ODCHYLENIE STANDARDOWE:**

```python
standard_deviation_data_1 = np.std(data_1, ddof=1)
standard_deviation_data_2 = np.std(data_2, ddof=1)

print(f"Odchylenie standardowe dla data_1 wynosi: {standard_deviation_data_1}")
print(f"Odchylenie standardowe dla data_2 wynosi: {standard_deviation_data_2}")
```

**9. ASYMETRIA ROZKŁADU:**

Rysujemy wykres:

```python
plt.hist(data_1, bins=20)
plt.axvline(x = mean_data_1, color='red', label = 'Średnia')
plt.axvline(x = median_data_1, color='black', label = 'Mediana')
plt.axvline(x = mode_data_1_int_value, color='yellow', label = 'Moda')
plt.legend(loc='upper right')
plt.title("Data 100")
plt.show()

plt.hist(data_2, bins=20)
plt.axvline(x = mean_data_2, color='red', label = 'Średnia')
plt.axvline(x = median_data_2, color='black', label = 'Mediana')
plt.axvline(x = mode_data_2_int_value, color='yellow', label = 'Moda')
plt.legend(loc='upper right')
plt.title("Data 10000")
plt.show()
```

Asymetria rozkładu nie występuję ponieważ mamy do czynienie z rozkładami symetrycznymi.

**10. SKALOWANIE ZMIENNEJ:**

10.1 Standaryzacja (technika skalowania)

```python
scaler_standardized = StandardScaler()

standardized_data_1 = scaler_standardized.fit_transform(data_1.reshape(-1, 1))

plt.hist(standardized_data_1, bins=20)
plt.axvline(x = standardized_data_1.mean(), color='red', label = 'Średnia')
plt.title("Ustandaryzowany zbiór data_1", size=15)
plt.show()

print(f'Średnia: {standardized_data_1.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_data_1, ddof=1)}')

###################################

standardized_data_2 = scaler_standardized.fit_transform(data_2.reshape(-1, 1))

plt.hist(standardized_data_2, bins=20)
plt.axvline(x = standardized_data_2.mean(), color='red', label = 'Średnia')
plt.title("Ustandaryzowany zbiór data_2", size=15)
plt.show()

print(f'Średnia: {standardized_data_2.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_data_2, ddof=1)}')
```

Wartości ze zbioru data_1 i data_2 zostały ustandaryzowane, czyli wartości zostały wyśrodkowane wobec średniej i wyskalowane wobec średniej.

10.2 Min Max Scaler (technika skalowania)

```python
scaler_MinMax = MinMaxScaler()

normalized_data_1 = scaler_MinMax.fit_transform(data_1.reshape(-1, 1))
plt.hist(normalized_data_1, bins=20)
plt.axvline(x = normalized_data_1.mean(), color='red', label = 'Średnia')
plt.title("Ustandaryzowany metodą MinMaxScaler zbiór data_1", size=13)
plt.show()
print(f'Średnia: {normalized_data_1.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_1, ddof=1)}')

#############################################

normalized_data_2 = scaler_MinMax.fit_transform(data_2.reshape(-1, 1))
plt.hist(normalized_data_2, bins=20)
plt.axvline(x = normalized_data_2.mean(), color='red', label = 'Średnia')
plt.title("Ustandaryzowany metodą MinMaxScaler zbiór data_2", size=13)
plt.show()
print(f'Średnia: {normalized_data_2.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_2, ddof=1)}')
```

Ustandaryzowano na wykresach wartości z bazy data_1 i data_2 dane motodą MinMaxScaler, czyli wszystkie wartości wyskalowano tak aby mieściły się w zakresie od ZERA do JEDNEGO.

**11. KORELACJA:**

Nie ma sensu badać korelacji dwóch zbiorów ze względu, że są to zbiory z wartościami losowymi. Mówimy o korelacji dodatniej lub ujemnej jeśli wraz ze wzrostem jednej wartości rośnie także druga w drugim zbiorze. Tutaj mamy wartości losowe, więc korelacja będzie bliska ZERu.

**WNIOSKI / różnice pomiędzy statystkami:**

Wartość średniej to suma wszystkich wartości zbioru podzielona przez ich liczbę. Średnia jest wrażliwa na wartości skrajne, które mogą znacząco zaburzyć jej wartość i przekłamać rzeczywisty obraz.
Mediana, w odróżnieniu od średniej, jest odporna na wartości odstające. Jest to wartość środkowa w uporządkowanym zbiorze danych. W przypadku parzystej liczby elementów medianą jest średnia z dwóch środkowych wartości. Zarówno średnia, jak i mediana, służą do określenia centralnej wartośco danych – czyli wartości dzielącej zbiór na dwie części.
Moda to wartość, która występuje w zbiorze najczęściej. Ma szczególne sens przy danych dyskretnych, gdzie można zaobserwować wyraźnie dominujące wartości.
Kolejnymi ważnymi miarami są kwartyle i zakres – obie opisują rozrzut danych:
•	Kwartyle dzielą uporządkowany zbiór na cztery równe części, z których każda zawiera około 25% obserwacji.
o	Pierwszy kwartyl (Q1) to wartość, poniżej której znajduje się 25% danych.
o	Drugi kwartyl (Q2) to mediana.
o	Trzeci kwartyl (Q3) obejmuje 75% danych.
•	Zakres to różnica między największą a najmniejszą wartością w zbiorze.
Rozstęp międzykwartylowy (IQR) to różnica między trzecim a pierwszym kwartylem (Q3 - Q1). Opisuje środkową połowę rozkładu danych i jest często wykorzystywany w wykresie pudełkowym (BoxPlot). 
Wariancja to miara opisująca, jak bardzo dane różnią się od średniej. Mierzy rozpiętość wartości w zbiorze – im mniejsza wariancja, tym dane są bardziej skoncentrowane wokół średniej.
Odchylenie standardowe (ang. standard deviation) to pierwiastek z wariancji i również opisuje rozproszenie danych wokół średniej. Większe odchylenie = dane są bardziej rozrzucone, mniejsze odchylenie = dane są bardziej skupione
Asymetria rozkładu (inaczej: skośność) to miara pokazująca, czy dane są symetryczne względem średniej, czy rozciągają się bardziej w jedną ze stron:
•	Skośność dodatnia – gdy średnia > mediana
•	Skośność ujemna – gdy średnia < mediana
•	Rozkład symetryczny – gdy średnia ≈ mediana
Skalowanie zmiennej to proces przekształcania danych wejściowych tak, aby znajdowały się w tym samym zakresie (np. od 0 do 1) lub miały porównywalne jednostki, co jest szczególnie ważne dla wielu algorytmów uczenia maszynowego.
Ostatnim ważnym elementem jest korelacja. Najczęściej stosuje się współczynnik korelacji Pearsona, który mierzy siłę i kierunek liniowej zależności między dwiema zmiennymi.
Wartość współczynnika mieści się w przedziale od –1 do 1:
•	–1 oznacza silną korelację ujemną,
•	0 brak korelacji,
•	1 silną korelację dodatnią.
